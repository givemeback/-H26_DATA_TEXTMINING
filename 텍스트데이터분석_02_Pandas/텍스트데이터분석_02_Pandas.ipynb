### 2. 시리즈 객체의 생성과 수정

```python
numbers =[1,2,3,4,5]
list_sr = pd.Series(numbers)
print(list_sr)
type(list_sr)
```

```python
print(list_sr.dtype, list_sr.shape, list_sr.size, list_sr.name)
print(list_sr.index)
print(list_sr.values)
```

```python
dict_data ={"a":1,"b":2,"c":3,"d":4,"e":5}
dict_sr = pd.Series(dict_data)
print(dict_sr.dtype, dict_sr.shape, dict_sr.size, dict_sr.name)
print(dict_sr.index)
print(dict_sr.values)
```

```python
sr1 = pd.Series(numbers, dtype=float)
print(sr1)
import numpy as np
sr2 = pd.Series(numbers, dtype=np.float32)
print(sr2)
```

```python
chars = ['a','b','c','d','e']
sr3 = pd.Series(numbers, index=chars, name='MySeries')
print(sr3.dtype, sr3.shape,sr3.size,sr3.name)
print(sr3.index)
print(sr3.values)
```

#### 3-2. 데이터프레임의 열(칼럼)

```python
score_dict = {
    '이름' : ['정약용','이순신','이율곡','홍길동'],
    '국어' : [70,100,90,85],
    '영어' : [80,80,95,100],
    '수학' : [100,70,90,90]
}

score_df = pd.DataFrame(score_dict)
score_df
```

```python
score_df['사회'] = [85,88,92,96]
score_df
```

```python
score_df['총점'] = score_df.국어 + score_df.영어 + score_df.수학 + score_df.사회
score_df['평균'] = score_df.총점/4
score_df
```

```python
x = [0,1,2,3,4,5]
y = [0,2,4,6,8,10]
df = pd.DataFrame()
df['x'] = x
df['y'] = y
df['x * y'] = df.x * df.y
df
```
